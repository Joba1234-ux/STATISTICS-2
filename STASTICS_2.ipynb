```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution>

def bernoulli_pdf(x, p):
        Parameters:
    - x: Outcome (0 or 1)
    - p: Probability of success (P(X=1))

    Returns:
    - Probability P(X=x)
    """
    if x not in [0, 1]:
        return 0  # Bernoulli distribution only defined for 0 and 1
    return p if x == 1 else 1 - p
```

```python
#19. Implement hypothesis testing using Z-statistics for a sample dataset

import numpy as np
from scipy.stats import norm
sample = np.array([10, 20, 30, 40, 45, 49])
sample_mean = np.mean(sample)
n = len(sample)
pop_mean = 50
pop_std = 2
z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z)))
print(f"Z-statistic: {z:.2f}")
print(f"P-value: {p_value:.4f}")
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject .")
```

```python
#20.  Create a confidence interval for a dataset using Python and interpret the result

import numpy as np
from scipy.stats import norm
data = np.array([5, 10, 15, 20, 25, 30])
mean = np.mean(data)
n = len(data)
pop_std = 2
confidence = 0.95
z = norm.ppf(1 - (1 - confidence) / 2)
margin = z * (pop_std / np.sqrt(n))
lower = mean - margin
upper = mean + margin
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python
#21. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

import numpy as np
from scipy import stats
data = np.random.normal(loc=100, scale=15, size=50)
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)
confidence = 0.95
t_crit = stats.t.ppf(1 - (1 - confidence)/2, df=n-1)
margin = t_crit * (sample_std / np.sqrt(n))
lower = sample_mean - margin
upper = sample_mean + margin
print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python
#28. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution

import numpy as np
from scipy.stats import skew, kurtosis
def analyze_distribution(n=1000, mean=0, std=1):
    data = np.random.normal(loc=mean, scale=std, size=n)
    skewness = skew(data)
    kurt = kurtosis(data)  # Excess kurtosis (0 for normal distribution)
    print(f"Skewness: {skewness:.4f}")
    print(f"Kurtosis (Excess): {kurt:.4f}")
analyze_distribution()
```

```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?

def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x : int (0 or 1)
    p : float (0 ≤ p ≤ 1), probability of success

    Returns:
    float : probability P(X = x)
    """
    if x not in (0, 1):
        return 0
    return p if x == 1 else 1 - p
```

```python
#9. Write a Python function to calculate Z-scores from a dataset and plot them

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):

    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std

    plt.plot(z_scores, 'o-', color='purple')
    plt.axhline(0, color='grey', linestyle='>>')
    plt.title('Z-scores of Dataset')
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.grid(True, arrowstyle='>>', alpha=0.5)
    plt.show()

    return z_scores
```

```python
#15. Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy.stats import norm

sample = [5, 10, 15, 20, 25]
pop_mean = 50
pop_std = 10

z = (np.mean(sample) - pop_mean) / (pop_std / np.sqrt(len(sample)))
p = 2 * (1 - norm.cdf(abs(z)))

print(f"Z = {z:.2f}, p = {p:.4f}")
print("Reject" if p < 0.05 else "Fail to reject")
```

```python
#16. Create a confidence interval for a dataset using Python and interpret the result

import numpy as np
from scipy.stats import norm
data = [5, 10, 15, 20, 25, 30]
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)
z = norm.ppf(0.975)
margin_error = z * (std / np.sqrt(n))
lower = mean - margin_error
upper = mean + margin_error
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python
#17.  Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

import numpy as np
from scipy.stats import t
data = np.random.normal(loc=5, scale=1, size=3)
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)
t_crit = t.ppf(0.975, df=n-1)
margin_error = t_crit * (std / np.sqrt(n))
lower = mean - margin_error
upper = mean + margin_error
print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python

                                            #PART 2

#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

import numpy as np
from scipy.stats import norm
sample = np.array([5, 7, 9, 11, 13, 15, 17])
population_mean = 30
population_std = 5
n = len(sample)
sample_mean = np.mean(sample)
z = (sample_mean - population_mean) / (population_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z)))
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z:.4f}")
print(f"P-value: {p_value:.4f}")
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis .")
else:
    print("Result: Fail to reject the null hypothesis .")



```

```python
#3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean

import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, population_mean, population_std):
    n = len(sample)
    sample_mean = np.mean(sample)

    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))
    p_value = 2 * (1 - norm.cdf(abs(z)))

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-score: {z:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < 0.05:
        print("Result: Reject the null hypothesis.")
    else:
        print("Result: Fail to reject the null hypothesis.")



```

```python
#4.  Perform a two-tailed Z-test using Python and visualize the decision region on a plot

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Data
sample = np.array([5, 7, 9, 11, 13, 15])
mu = 50
sigma = 5
n = len(sample)
sample_mean = np.mean(sample)
se = sigma / np.sqrt(n)
z = (sample_mean - mu) / se
p = 2 * (1 - stats.norm.cdf(abs(z)))
z_crit = stats.norm.ppf(0.975)


print(f"Z = {z:.2f}, p = {p:.4f}, Z_crit = ±{z_crit:.2f}")
x = np.linspace(-4, 4, 100)
y = stats.norm.pdf(x)
plt.plot(x, y)
plt.fill_between(x, y, where=(x <= -z_crit) | (x >= z_crit), color='red', alpha=0.3)
plt.axvline(z, color='black', linestyle='--', label=f'Z = {z:.2f}')
plt.legend()
plt.title("Two-tailed Z-test")
plt.show()
```

```python
#6. Write a Python program to perform an independent T-test and interpret the results


import numpy as np
from scipy import stats
group1 = np.array([5, 10, 15, 20, 25])
group2 = np.array([2, 4, 6, 8, 10])
t_stat, p_value = stats.ttest_ind(group1, group2)
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.4f}")
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis .")
else:
    print("Result: Fail to reject the null hypothesis ")
```

```python
#7.  Perform a paired sample T-test using Python and visualize the comparison results

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
before = np.array([5, 10, 15, 20, 25])
after  = np.array([2, 4, 6, 8, 10])
t_stat, p_value = stats.ttest_rel(before, after)
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.4f}")
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis .")
else:
    print("Result: Fail to reject the null hypothesis.")
x = np.arange(len(before))
plt.figure(figsize=(8, 5))
plt.plot(x, before, marker='o', label='Before', linestyle='--', color='blue')
plt.plot(x, after, marker='s', label='After', linestyle='--', color='green')
plt.title("Paired Sample Comparison")
plt.xlabel("Subject Index")
plt.ylabel("Scores")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
